```python
s=int(input("eneter the sixe of queue"))
queue=["NULL"]*s
rear=front=-1
def enqueue(X):
    global rear
    global front
    if s-1 !=rear:
        if rear==-1:
            rear+=1
            front+=1
            queue[rear]=X
        else:
            rear+=1
            queue[rear]=X
    else:
        print("QUEUE OVERFLOW")
def dequeue():
    global rear
    global front
    if front == -1:
        print("queue is empty")
    elif front == s:
        print("queue is full")
    else:
        temp =queue[front]
        print("DELETED :- ",temp)
        queue[front]="NULL"
        front+=1
        
```

```python
enqueue(4)
print(queue)
enqueue(4)
print(queue)

enqueue(4)
print(queue)
```
